```python
import string
from collections import Counter

with open('/content/drive/MyDrive/Colab Notebooks/Avengers.txt', 'r') as file:
    avengers = file.read()

temp1 = avengers.lower().split()
# print(temp1)
temp2 = []
for s in temp1:
    temp2.append(s.replace("'","").replace('-','').replace('...','').strip(string.punctuation))
        #.strip메서드는 문자열 중간에 있는 ...을 지우지 못함

# result = ' '.join(temp2)

# print(temp2, '\n')

temp3 = Counter(temp2)
# temp3 = dict(sorted(Counter(temp2).items(), reverse=True))
temp3 = dict(sorted(Counter(temp2).items(), reverse=True))
# print(temp3, '\n')

temp4 = dict(enumerate(temp3))
print(temp4, '\n')

input_val = input("문장을 입력해주세요: ")


```
